##Лабораторная работа №7.

Программа, которая:
1. Запрашивает у пользователя цитату (одну строку текста).
2. Анализирует текст посимвольно с помощью цикла for.
3. Проверяет, стоит ли пробел после знаков препинания ., ,, !, ?.
4. Если обнаружено нарушение программа:
- Автоматически исправляет текст: вставляет пробел после знака препинания.
- Выводит исправленную цитату.
- Выводит предупреждение:
- ⚠️ Внимание: обнаружены пропущенные пробелы после знаков препинания и
исправлены.
5. Если нарушений нет — выводит текст без изменений и не показывает предупреждение.

```python
# Запрашиваем цитату у пользователя
quote = input("Введите цитату: ")

# Список знаков препинания, после которых должен быть пробел
punctuation_marks = ['.', ',', '!', '?', ':', ';', '—', '-']

# Переменные для отслеживания состояния
corrected_text = ""
has_errors = False
i = 0

# Анализируем текст посимвольно
while i < len(quote):
    current_char = quote[i]
    corrected_text += current_char

    # Проверяем, является ли текущий символ знаком препинания
    if current_char in punctuation_marks:
        # Проверяем, есть ли следующий символ и не является ли он пробелом
        if i + 1 < len(quote) and quote[i + 1] != ' ':
            # Добавляем пробел после знака препинания
            corrected_text += ' '
            has_errors = True

    i += 1

# Выводим результат
print("\nИсправленная цитата:")
print(corrected_text)

# Выводим предупреждение, если были найдены ошибки
if has_errors:
    print("\n⚠️ Внимание: обнаружены и исправлены пропущенные пробелы после знаков препинания.")
else:
    print("\n✓✅ Ошибок не обнаружено.")
```
