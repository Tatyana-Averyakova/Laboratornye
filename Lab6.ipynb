##Лабораторная работа №6. Анализ эмоциональной окраски постов.

Программа для анализа эмоциональной окраски постов.

- найти все слова, написанные с заглавными буквами;
- посчитать сколько раз встречается такое слово;
- однобуквенные слова и слова, содержащие цифры исключить из анализа.

```python
# определить, что такое "заглавная буква"
def is_uppercase_letter(ch):
    return ch in "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ"

def is_letter(ch):
    return ch in "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯабвгдеёжзийклмнопрстуфхцчшщъыьэюя"

# удалить знаки препинания и заменить их на пробелы
def clean_text(txt):
  punctuation = ".,:;!?«»\"'-~=>#$—)()"
  cleaned = ""
  for char in txt:
    if char in punctuation:
      cleaned += " "
    else:
      cleaned += char
  return cleaned

# счет количества слов
def count_words(capital_l):
  name_counts = {}
  for name in capital_l:
    if name in name_counts:
      name_counts[name] += 1
    else:
      name_counts[name] = 1
  return name_counts

#текст
text = """
«СРОЧНО! Это ВАЖНО! Все должны это видеть (а тем кто НЕ видел, СРОЧНО расскажите!!!111). НЕ
игнорируйте — это РЕАЛЬНО спасёт жизни!
~СМС пришло в 3 часа. => Я был в ШОКЕ. ПОЖАЛУЙСТА, $$поделитесь$$! #ВАЖНО #СПАСИТЕ
P.S. Я не шучу. ПОВТОРЯЮ: ЭТО ВАЖНО!»
"""

#текст разбить на слова
cl_text = clean_text(text)
print(cl_text)
words = cl_text.split()
print(words)

# все слова, которые:
# написанные с заглавной буквы;
# слова состоят из букв;
# длина слов > 1

capital_l = []
for word in words:
    if len(word) > 1 and is_uppercase_letter(word[0]) and all(is_letter(ch) for ch in word):
        capital_l.append(word)


# счет слов
name_counts = count_words(capital_l)

# вывод
print("Слова с заглавными буквами и их количество:")
for name, count in name_counts.items():
    print(f"{name}: {count}")
```
