##Лабораторная работа №5. Программа заполнения билета для регистрации участника на конференцию.


**Данные участника:**
- Фамилия - не должна быть пустой, минимум две буквы;
- Имя - не должно быть пустым, минимум две буквы;
- Возраст - число. От 18 до 90;
- Формат участия - онлайн или оффлайн.

После выводится билет с данными участника.

```python
#входные параметры

class Participant:
    def __init__(self, last_name, first_name, age, participation_format):
        self.last_name = last_name
        self.first_name = first_name
        self.age = age
        self.participation_format = participation_format

    #вывод данных участника
    def display_ticket(self):
        print("\n" + "="*50)
        print("          БИЛЕТ УЧАСТНИКА КОНФЕРЕНЦИИ")
        print("="*50)
        print(f"Фамилия:           {self.last_name}")
        print(f"Имя:               {self.first_name}")
        print(f"Возраст:           {self.age} лет")
        print(f"Формат участия:    {self.participation_format}")
        print("="*50)
        print("Благодарим за регистрацию! Ждем Вас на конференции!")
        print("="*50)

#ввод и проверка "Фамиилия"
def validate_last_name(last_name):
    """Проверка фамилии"""
    if not last_name:
        return "Фамилия не может быть пустой"
    if len(last_name) < 2:
        return "Фамилия должна содержать минимум 2 буквы"
    if not last_name.isalpha():
        return "Фамилия должна содержать только буквы"
    return None

#ввод и проверка "Имя"
def validate_first_name(first_name):
    """Проверка имени"""
    if not first_name:
        return "Имя не может быть пустым"
    if len(first_name) < 2:
        return "Имя должно содержать минимум 2 буквы"
    if not first_name.isalpha():
        return "Имя должно содержать только буквы"
    return None

#ввод и проверка "Возраст"
def validate_age(age_str):
    """Проверка возраста"""
    try:
        age = int(age_str)
        if age < 18:
            return "Возраст должен быть не менее 18 лет"
        if age > 90:
            return "Возраст должен быть не более 90 лет"
        return None
    except ValueError:
        return "Возраст должен быть числом"

#ввод и проверка "Формат участия"
def validate_participation_format(format_str):
    """Проверка формата участия"""
    format_str = format_str.lower()
    if format_str in ["онлайн", "online"]:
        return "онлайн", None
    elif format_str in ["оффлайн", "офлайн", "offline"]:
        return "оффлайн", None
    else:
        return None, "Формат участия должен быть 'онлайн' или 'оффлайн'"

def get_valid_input(prompt, validation_func, is_participation_format=False):
    """Получение корректного ввода от пользователя"""
    while True:
        user_input = input(prompt).strip()

        if is_participation_format:
            validated_input, error = validation_func(user_input)
            if error:
                print(f"Ошибка: {error}")
                print("Доступные варианты: онлайн, оффлайн")
            else:
                return validated_input
        else:
            error = validation_func(user_input)
            if error:
                print(f"Ошибка: {error}")
            else:
                return user_input

def main():
    print("РЕГИСТРАЦИЯ УЧАСТНИКА КОНФЕРЕНЦИИ")
    print("=" * 40)

    # Ввод данных с валидацией
    last_name = get_valid_input("Введите фамилию: ", validate_last_name)
    first_name = get_valid_input("Введите имя: ", validate_first_name)

    age_input = get_valid_input("Введите возраст: ", validate_age)
    age = int(age_input)

    participation_format = get_valid_input(
        "Введите формат участия (онлайн/оффлайн): ",
        validate_participation_format,
        True
    )

    # Создание участника и вывод билета
    participant = Participant(last_name, first_name, age, participation_format)
    participant.display_ticket()

if __name__ == "__main__":
    main()
```
